```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
raw_data = np.loadtxt('data/unemployment.csv', 
           delimiter=',', usecols=(0,1), 
           unpack=True, dtype=float, skiprows=1)

print(list(raw_data))
```

```python
x_data = raw_data [0]
y_data = raw_data [1]

# Drawing scatter plot
plt.plot(x_data, y_data, "ro")
plt.show()
```

```python
def least_squares_fit(x,y):
    """given training values for x and y,
    find the least-squares values of alpha and beta"""
    beta = np.dot((x-mean(x)), (y-mean(y)))  / sum((x-mean(x)) ** 2)
    alpha = mean(y) - beta * mean(x)
    return alpha, beta

alpha, beta = least_squares_fit(x_data, y_data)
print("alpha", alpha)
print("beta", beta)
```

```python
# Y = alpha + beta * X
# Drawing scatter plot

plt.plot(x_data, y_data, "ro")
plt.plot(x_data, alpha + beta * x_data)
plt.show()
```
